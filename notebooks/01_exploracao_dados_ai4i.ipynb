```python
#Célula 2: Importação das Bibliotecas
import pandas as pd
import numpy as np

#Configuração para visualização no Pandas
pd.set_option('display.max_columns', None)

print("Bibliotecas importadas com sucesso!")
```

```python
#Célula 3: Carregamento do Dataset AI4I 2020
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00601/ai4i2020.csv"
df = pd.read_csv(url)

#Visualizando as primeiras linhas
df.head()
```

```python
#Célula 4: Limpeza e Preparação para o Contexto Industrial

#1. Remover colunas de identificação (não são sensores, não ajudam na física do problema)
df_clean = df.drop(['UDI', 'Product ID'], axis=1)

#2. Tratamento de Vazamento de Dados (Data Leakage)
cols_to_drop = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']
df_clean = df_clean.drop(cols_to_drop, axis=1)
```

```python
#Célula 5: Tradução para Human-Centricity

mapa_nomes = {
    'Type': 'Tipo do Equipamento',
    'Air temperature [K]': 'Temperatura do Ar (K)',
    'Process temperature [K]': 'Temperatura do Processo (K)',
    'Rotational speed [rpm]': 'Rotacao (rpm)',
    'Torque [Nm]': 'Torque (Nm)',
    'Tool wear [min]': 'Desgaste da Ferramenta (min)',
    'Machine failure': 'Falha'
}

df_clean = df_clean.rename(columns=mapa_nomes)
```

```python
#Célula 6: Separação X (Features) e y (Target)
y = df_clean['Falha']
X = df_clean.drop('Falha', axis=1)

#Verificação de como ficaram os dados
print("Dataset pronto para Modelagem:")
display(X.head())
```

```python
#Célula 7:Salvar os dados processados para o próximo notebook
X.to_csv('../data/processed/X_industrial.csv', index=False)
y.to_csv('../data/processed/y_industrial.csv', index=False)
```
